```python
import librosa

import matplotlib.pyplot as plt

from scipy import signal
from scipy.io import wavfile

import numpy as np

%matplotlib inline
```

```python
sample_rate, samples = wavfile.read('chromo.wav')
samples = np.array(samples[:, 0], dtype=float)
```

```python
ans = np.abs(librosa.core.cqt(samples, sample_rate, hop_length=2048, fmin=216.626, n_bins=20, bins_per_octave=12))
```

```python
plt.pcolormesh(ans)
```

```python
sample_rate, samples = wavfile.read('BachCDur.wav')
samples = np.array(samples[:, 0], dtype=float)[:sample_rate * 10]
ans = np.abs(librosa.core.cqt(samples, sample_rate, hop_length=2 ** 12, fmin=216.626, n_bins=20, bins_per_octave=12))
```

```python
plt.pcolormesh(ans)
```
