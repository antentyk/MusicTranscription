```python
from scipy import signal
from scipy.io import wavfile
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sample_rate, samples = wavfile.read('chromo.wav')
samples = samples[:, 0]
plt.figure()
plt.plot(samples[:200])
plt.figure()
plt.plot(samples)
```

```python
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate, nperseg=256)
frequencies
```

```python
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate, nperseg=1024)
frequencies
```

```python
c1line = 250
c2line = 555
```

```python
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate, nperseg=2048)
spectrogram = spectrogram[(frequencies > c1line) & (frequencies < c2line),:]
frequencies = frequencies[(frequencies > c1line) & (frequencies < c2line)]
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.pcolormesh(times, frequencies, spectrogram)
```

```python
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate, nperseg=256)
spectrogram = spectrogram[(frequencies > c1line) & (frequencies < c2line),:]
frequencies = frequencies[(frequencies > c1line) & (frequencies < c2line)]
plt.pcolormesh(times, frequencies, spectrogram)
print(frequencies)
```
